```python
while True:
    idade = int(input("Qual é sua idade? "))
    valor_total = float(input("Qual valor total você tem na carteira? "))

    if idade < 16:
        print("Entrada permitida apenas com acompanhante")
    elif idade >= 16 and valor_total >= 30:
        print("Ingresso comprado! Bom filme.")
    else:
        print("Saldo insuficiente para o ingresso.")

    continuar = input("Deseja verificar outro cliente? (s/n): ")

    if continuar == 'n' or continuar == 'N':
        print("Encerrando o programa...")
        break
```
